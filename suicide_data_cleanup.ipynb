```python
import pandas as pd
import os
```

```python
filepath = os.path.join('.', 'suicide_original_data.csv')
suicide_df = pd.read_csv(filepath)
suicide_df.head()
```

```python
suicide_df['age'] = suicide_df['generation'].str.cat(suicide_df['age'], sep =": ")
suicide_df = suicide_df.drop(columns=['generation', 'HDI for year', 'country-year'])
suicide_df.head(10)
```

```python
# get names of indexes for which 
# column Age has value 21 
index_names = suicide_df[suicide_df['year'] != 2015].index 

# drop these row indexes 
# from dataFrame 
suicide_2015 = suicide_df.drop(index_names, inplace = False)
suicide_2015 = suicide_2015.rename(columns = {' gdp_for_year ($) ': 'gdp_for_year', 'gdp_per_capita ($)': 'gdp_per_capita'})

suicide_2015.head(20)
```

```python
suicide_2015.set_index('country', inplace=True)
```

```python
suicide_2015.to_csv(r'suicide_2015.csv', index = False)
```

```python
suicide_2015_summary = suicide_2015.drop(columns=['year', 'sex', 'age', 'suicides/100k pop'])
suicide_2015_summary.head()
```

```python
suicide_2015_summary.dtypes
```

```python
# suicide_2015_summary.gdp_for_year = suicide_2015_summary.gdp_for_year.astype('float64') 
# suicide_2015_summary.dtypes
```

```python
suicide_2015_group = suicide_2015_summary.groupby(['country', 'gdp_for_year', 'gdp_per_capita']).sum()
suicide_2015_group.head(20)
```

```python
suicide_2015_summary = pd.DataFrame(suicide_2015_group.reset_index())
suicide_2015_summary.head(10)
```

```python
suicide_2015_summary.set_index("country", inplace=True)

suicide_2015_summary.head()
```
